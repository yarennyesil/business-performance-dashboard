```python
# Business Performance Visualization Dashboard

## Projenin Amacı

Bu projenin amacı, bir şirketin satış verilerini Matplotlib kullanarak görselleştirmek ve verileri daha anlaşılır hale getirmektir.

Ham veriler çoğu zaman yöneticiler için anlamlı olmayabilir. Bu nedenle veri görselleştirme teknikleri kullanılarak satış performansı kolayca analiz edilebilir.

## Veri Görselleştirmenin Önemi

Veri görselleştirme:

- Trendleri hızlı görmemizi sağlar
- Karşılaştırma yapmayı kolaylaştırır
- Yönetim kararlarını destekler
- Karmaşık verileri basitleştirir

## Kullanılan Matplotlib Konuları

Bu projede aşağıdaki Matplotlib konuları kullanılmıştır:

- Matplotlib giriş
- Line plot (çizgi grafiği)
- Bar plot (sütun grafiği)
- Grafik özelleştirme
- Subplots (iç içe grafikler)

Bu sayede satış verileri bir dashboard mantığında sunulacaktır.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Aylar
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Aylık satışlar
monthly_sales = [12000, 15000, 17000, 14000, 18000, 21000,
                 25000, 24000, 23000, 26000, 30000, 32000]

# Ürün bazlı satışlar
product_sales = {
    "Laptop": 55000,
    "Phone": 42000,
    "Tablet": 18000,
    "Accessories": 12000
}

# Bölge bazlı satış
region_sales = {
    "Europe": 40000,
    "Asia": 35000,
    "America": 30000,
    "Middle East": 20000
}

print("Veriler hazırlandı.")
```

```python
plt.figure(figsize=(10,5))

plt.plot(months, monthly_sales, marker="o")

plt.title("Monthly Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales ($)")

plt.grid(True)

plt.show()
```

```python
products = list(product_sales.keys())
sales = list(product_sales.values())

plt.figure(figsize=(8,5))

plt.bar(products, sales, color=["blue","green","orange","purple"])

plt.title("Product Sales Comparison")
plt.xlabel("Products")
plt.ylabel("Sales ($)")

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(months, monthly_sales,
         color="red",
         linestyle="--",
         marker="o",
         label="Monthly Sales")

plt.title("Customized Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales ($)")

plt.legend()
plt.grid(True)

plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize=(12,8))

# 1️⃣ Aylık satış trendi
axs[0,0].plot(months, monthly_sales, marker="o")
axs[0,0].set_title("Monthly Sales Trend")

# 2️⃣ Ürün satışları
axs[0,1].bar(product_sales.keys(), product_sales.values())
axs[0,1].set_title("Product Sales")

# 3️⃣ Bölge satışları
axs[1,0].bar(region_sales.keys(), region_sales.values())
axs[1,0].set_title("Regional Performance")

# 4️⃣ Genel satış ortalaması
avg_sales = np.mean(monthly_sales)
axs[1,1].bar(["Average Sales"], [avg_sales])
axs[1,1].set_title("Average Monthly Sales")

plt.tight_layout()

plt.show()
```
